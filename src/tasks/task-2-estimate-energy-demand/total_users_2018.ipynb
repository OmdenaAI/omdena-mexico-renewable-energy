```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('users_consumption_2018.csv', skiprows=2)
data.head()
```

```python
data.shape
```

```python
# Transform variables to categories (factors)
data['Tarifa'] = data ['Tarifa'].astype('category')
data['Municipio'] = data['Municipio'].astype('category')
```

```python
data.info()
```

```python
data.columns
```

```python
# Delete columns 
data = data.drop(['Cve Edo', 'Cve  Inegi', 'Cve  Mun'], axis=1)
```

```python
# The numercial columns are as objects, each data python read it as string, so I deleted commas and transformed it in 
# floats
data = data.replace({',': ''}, regex = True)
data.iloc[:, 3:] = data.iloc[:, 3:].astype(float)
```

```python
# data['Abril'] = data['Abril'].str.replace(',', '').astype(float)
```

```python
data.info()
```

```python
data.head()
```

```python
# Sum the months for a total of 2018
data['sum'] = data.iloc[:, 2:].sum(axis = 1)
```

```python
total_2018 = data[['Entidad Federativa','Municipio', 'Tarifa', 'sum']]
total_2018.head()
```

```python
total_2018.to_csv('total_2018.csv')
```
