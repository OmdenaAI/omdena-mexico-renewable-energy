Function for cleaning all the dataset

```python
def data_cleaning(data):
  data = data.iloc[2:, :4]
  data = data[(data['Localidades/Población'] != 'Localidad')]
  data = data.drop(columns='Localidades/Población')
  data = data.rename(columns={'Total de localidades y población1': 'poblacion'})

  # Separate key of each municipio
  new = data['Municipio'].str.split(' ', n = 1, expand = True)
  data['key_municipio'] = new[0]
  data['municipio'] = new[1]
  data.drop(columns = 'Municipio', inplace = True)

  # Separate key of each state
  new2 = data['Entidad federativa'].str.split(' ', n = 1, expand = True)
  data['key_state'] = new2[0]
  data['entidad_federativa'] = new2[1]
  data.drop(columns = 'Entidad federativa', inplace = True)

  # Unite keys
  data['key'] = data['key_state'] + data['key_municipio']
  data.drop(columns = ['key_state', 'key_municipio'], inplace = True)

  return data
```

```python
import glob
import os 
os.chdir('/content')

extension = 'xlsx'
all_files = [i for i in glob.glob('*.{}'.format(extension))]

files = []
for filename in all_files:
  data = pd.read_excel(filename, skiprows=5)
  files.append(data)
```

```python
files_clean = []
for i in files:
  data_clean = data_cleaning(i)
  files_clean.append(data_clean)
```

```python
pop_2020 = pd.concat(files_clean)
pop_2020.info()
```
